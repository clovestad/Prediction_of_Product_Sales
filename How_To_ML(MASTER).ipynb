#Types of Features
Our datasets will likely contain a variety of datatypes, but ultimately, these need to be converted into numerical values so they can be interpreted by the computer. Different types of features will need a different type of preprocessing.

- Three Types of Features

Numeric Features are integers or floats that represent a quantity. They can also be referred to as quantitative features.

Numeric features should be integers or floats, but if the data type indicates "object", it could be due to the presence of a non-numeric character such as the dollar sign: $3.00. Be sure to address these characters and change the data type to a numeric format.

Examples:

Price
mpg
Number of rooms

- Ordinal Features are categories that represent different classes and that have a distinct ordering.

They can be strings or integers if the integer represents an ordered class.

Examples:

low, medium, high
One star, two stars, three stars, four stars, and five stars
0,1,2,3 where 0 is small, 1 is medium, 2 is large, and 3 is extra large

- Categorical Features (nominal) are categories that represent different classes. They are not ordered.

They can be strings or integers if the integer represents a nonordered class.

Examples:

male, female, nonbinary
red, green, blue, yellow, purple, green
0, 1, 2, 3 where 0 is single, 1 is married, 2 is divorced, and 3 is widowed

- We will now look at how each of these types of features is transformed in preparation for machine learning,

- Transforming Features for Machine Learning
In order for features to be interpreted by a machine learning algorithm, the data must be in a numeric form (integers or floats).

#Transforming Numeric Features

- These are already in a numeric format so can be used without further transformation. However, it is very common and sometimes required, to scale numerical features. Scaling is discussed in a separate lesson.

- Transforming Ordinal Features

- If they are strings, they will need to be mapped to numeric values representing the order of the classes.

- Note that sometimes this mapping has already occurred such as in the example above where 0 is small, 1 is medium, 2 is large, and 3 is extra large. In this case, no further transformation is needed.

- Examples of ordinal string transformation:

  - low, medium, and high will be transformed to 0, 1, 2
One star, two stars, three stars, four stars, and five stars will be transformed to 0, 1, 2, 3, 4

##Transforming Categorical (Nominal) Features
You need to be careful when your categories are unordered. It is not sufficient to simply map each class to a number as can be done with ordinal features.

This is because machine learning models will interpret higher numbers as having a higher value than lower numbers. However, there should not be a higher or lower value associated with nominal variables. For example, if you just replaced the values red, green, and blue with the numbers 0, 1, and 2, the machine learning model would interpret blue as being higher/more than red and green. We know that this should not be the case; all of these colors should be treated as equal values since no one color is inherently better/higher than another.

To deal with this, we can one-hot encode our categories. What this does is creates a binary column for each class in the column.

Note that the Bldg Type column in our dataset is an example of a nominal feature.

```python
# View values in the Bldg Type column
X_train['Bldg Type'].value_counts(dropna=False)

```

One hot encoding will create a binary column for each Bldg Type in our dataset.


Notice, in each of the rows, one (and only one) of the building types is "hot" (assigned 1). This is essentially just a bunch of binary columns that represent whether or not the value in that row is that Bldg Type.

For example, the first row represents the Bldg Type category: "1Fam". So, it has a "1" in the "Bldg Type_1Fam" column (signifying the presence of the "1Fam" Bldg Type) and a "0" in the other one-hot-encoded categories: "Bldg Type_2fmCon", "Bldg Type_Duplex, " etc.

#Imputation Strategies
Simple imputer can use any of 5 strategies to impute values:

- 'mean', fill missing values with the mean average of the column they are in.
- 'median', fill missing values with the median average of the column they are in.
- 'mode', fill missing values with the mode average of the column they are in.
- 'most_frequent', fill missing values with the most frequent value in the column they are in (equivalent to 'mode' for numeric columns).
- 'constant', provide a constant value to use to fill in missing values. A common choice for categorical data is 'missing'.
##Simple Imputer
Before working with the simple imputer, you should have already:

- Performed the train-test-split
There are 4 main steps to imputation to prevent data leakage:

1. Define a list of features to use with a particular imputer.

2. Instantiate the imputer with the desired imputation strategy.

3. Fit the imputer. The simple imputer will learn the values it needs to impute from ONLY the training data. This step is accomplished using the imputer.fit(X_train) method.

- Fit once on training data only
4. Transform the values. In the final step, the knowledge learned the first step will be used to transform the data. In this case, .transform will impute the missing values with the median values calculated in the fit step.

4.a() Transform training data
4.(b) Transform testing data

#Ordinal and One-Hot Encoders

##What is One-Hot Encoding for Nominal Features?
When words are not ordered, such as "red", "blue", or "green," we can one-hot encode the data. First, one-hot encoding does not capture the meaning of the words. The computer does not know what blue looks like, but it can still find relationships between the color and other variables in the context of a dataset. In order to represent non-ordered, or 'nominal' features, we do the following:

1. Create a new column for every category present in the feature.

2. Set the value of each of the new columns to 1 if that row corresponds to the original category.

3. Set the value of each of the new columns to 0 if they do not.

4. Remove the original column.

#One-Hot Encoding
- OneHotEncoder follows the same process as our other transformers:

1. Define a list of nominal features to one-hot encode

2. Instantiate the encoder

3. Fit the OneHotEncoder ONLY on the categorical training data.

4. Transform the categorical training data to a one-hot encoded form.

5. Transform the categorical testing data to a one-hot encoded form.

1. Define a list of nominal Features

We only want to one-hot encode the categorical features that have not been encoded ordinally. We will make a new list of columns to encode using OneHotEncoder, which we will call ohe_cols.

We know that we have already encoded the ordinal features, so any remaining categorical feature will be encoded using OneHotEncoder.  We will define our list of features for one-hot encoding by dropping the ordinal columns.

```python
# saving list of categorical features to one-hot-encode
ohe_cols = X_train_cat_imputed.drop(columns=ordinal_cols).columns
ohe_cols
```

2. Instantiate the Encoder

Scikit-Learn's OneHotEncoder class will do all of the one-hot encoding work for us, but we want to make some changes to the default settings when we instantiate the OneHotEncoder.

sparse_output = False

By default, it will return what is called a 'sparse matrix'. This is a form of data compression used when arrays are mostly filled with 0s. Instead of keeping track of so many 0s, the compressed version just has information about where the data is NOT 0. This is great for saving memory! But not great if we want to use the data in a new dataframe. We can get OneHotEncoder to return a normal array that is compatible with pandas, called a Dense array by specifying sparse=False inside our OneHotEncoder.

handle_unknown ='ignore'

By default, OneHotEncoder will throw an error if we try to transform data that has categories that were not present in the data it was fitted on and does not have columns for. We can pass the argument 'handle_unknown='ignore'' to tell it to ignore any categories that were not present during the fit step (values it encounters in the test set that were not present in the train set). In this case, all columns corresponding to that feature will have the value 0.



```python
# Instantiate one hot encoder
ohe_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
# Fit the OneHotEncoder on the training data
ohe_encoder.fit(X_train_cat_imputed[ohe_cols])
```

#Standarization

##Why Standardize or Scale Data?
A common assumption in many types of models is that your data is scaled appropriately.

Terms to Know:
Scale
Generally means to change the range of the values. The shape of the distribution doesnt change. Think about how a scale model of a building has the same proportions as the original, just smaller. Thats why we say it is "drawn to scale".

Standardize
Standardizing is one of several kinds of scaling. It means scaling the values so that the distribution has a standard deviation of 1 with a mean of 0. It outputs something very close to a normal distribution.

Note: Scaled values lose their original units. Dollars are no longer in dollar units, meters are no longer in meter units, etc.

The Math
Standardization is calculated as:

standardized_feature = (feature - mean_of_feature) / std_dev_of_feature

##Standardizing Data in Python

In Python, you can scale data by using Scikit-learn's StandardScaler.

To avoid data leakage, the scaler should fit on only the training set. When the scaler fits on data it calculates the means and standard deviations of each feature. Then the scaler can be used to transform both the training and test sets based on the calculations done during the fit step. This means that the average and variance (standard deviation) will be calculated using only the training data because we want to keep information in the test data, including information about means and variances, reserved for only the final model evaluation. The scaling of target values (y) is generally not required.

#Pipelines

A pipeline contains multiple transformers (or even models!) and performs operations on data IN SEQUENCE. When a pipeline is fit on data, all of the transformers inside it are fit. When data is transformed using a pipeline, the data is transformed by the first transformer first, the second transformer second, etc. A pipeline can contain any number of transformers as long as they have .fit() and .transform() methods. These are called 'steps'.

NOTE: The pipeline will REPLACE the individual transformation steps we have demonstrated in the previous lessons! You can choose to do these steps separately OR you can use a pipeline, but you do NOT do them individually and use the pipeline! Thus we will return to our non-processed data to start this demonstration.

##Why Should I Use Pipelines for Machine Learning?
Reasons to use pipelines:

1. Pipelines use less code than doing each transformer individually. Since each transformer is fit in a single .fit() call, and the data is transformed by all of the transformers in the pipeline in a single .transform() call, pipelines use significantly less code.
2. Pipelines make your preprocessing workflow easier to understand. By reducing the code and displaying a diagram of the pipeline you can show your readers clearly how your data is being transformed before modeling.
3. Pipelines are easy to use in production models. When you are ready to deploy your model to use in new data, a preprocessing pipeline can ensure that new data can be quickly and easily preprocessed for modeling.
Pipelines can prevent data leakage.
4. Pipelines are designed to only be fit on training data. Later you will learn a technique called 'cross-validation' and pipelines will simplify performing this without leaking data.

#Column Transformer

Real-world data often contain heterogeneous data types. When processing the data before applying the final prediction model, we typically want to use different preprocessing steps and transformations for those different types of columns.

A simple example: we may want to scale the numerical features and one-hot encode the categorical features.

In previous lessons, you transformed groups of columns separately and then stitched them together manually. In this lesson, you will use ColumnTransformer to do that work for you.

column-transformer.png

You can see here that the column transformer works in parallel. It scales the numeric data, one hot encodes the categorical data and everything else that is not one of those categories is passed through to the final dataset unchanged. These steps occur at the same time.

Also, notice that the categorical columns are now larger than they were before the column transformer. That is because when one-hot encoding, there is a separate column created for each category in each categorical column.

Using a column transformer lets you apply different types of transformers to different columns in your data. It is convenient and makes it easy to see which transformations were applied to which columns during the pre-processing phase. While it is not necessary, it is very common to use a pipeline to define one or more pathways. We will demonstrate the use of pipelines in our column transformers. (To see an example of an individual transformer used in a pathway, refer to the FAQs at the bottom of the page).

## Step 7- Evaluate the Model Using Regression Metrics


In order to determine if a model is good or not, we will need metrics that inform us of how wrong our model's predictions were.

The difference between the predicted target and the true target is called the error (A.K.A. "residual").

image.png

In order to assess our model, we need a single number that can summarize how wrong our model was overall.

One approach we could take is to sum up all of the errors for every prediction:

- To sum up all of the errors in math notation, we add the summation symbol (sigma) and an "i" to indicate each individual data p

-However, there's a major problem with this!

Our model can be wrong in both directions; it can guess too high or too low, creating both positive and negative errors.

If we simply summed all of the errors, we would have positive and negative errors cancel out. For example, an error of +5000 (over) and an error of -5000 (under) added together is zero, which would make it seem like there were no errors at all!

Instead, we need a metric that can add all of the errors without considering if the error is positive vs. negative, but there are multiple ways to do so.

Mean Absolute Error (MAE)
Mean absolute error measures the average of the absolute values of all of the errors our model makes.

By using absolute error, we prevent errors in one direction from "canceling" out errors in the opposite direction.

For MAE, we dont care if the prediction was too high or too low, we are just concerned with how large it is (the magnitude).

By using absolute value, our errors in either direction would accumulate to 10,000 which is a better representation of our model's performance.

For this metric lower is better because we want lower errors.

## MAE in Python
We can calculate the mean absolute error using NumPy:

```python
# Calculating MAE with numpy
train_MAE = np.mean(np.abs(y_pred_train - y_train))
test_MAE = np.mean(np.abs(y_pred_test - y_test))
print(f'Model Training MAE: {train_MAE:,.2f}')
print(f'Model Testing MAE: {test_MAE:,.2f}')
```

Note: we have added a format code to our f-string print statements.

By adding :,.2f inside the curly brackets in an f-string, we are telling Python to:

- Use a comma to separate thousands.
- Use 2 decimal places for floats.
Or, we can use the Scikit-Learn metrics function.

```python
# Calculating MAE with sklearn
from sklearn.metrics import mean_absolute_error
train_MAE = mean_absolute_error(y_train, y_pred_train)
test_MAE = mean_absolute_error(y_test, y_pred_test)
print(f'Model Training MAE: {train_MAE:,.2f}')
print(f'Model Testing MAE: {test_MAE:,.2f}')
```

Both approaches give us the same number. This result means that, on average, our model was off on its predictions by $21,158 for the training set and $19,928 for the test set.

Since the median house price in this dataset is $159,900, our model is off by about 13% on average.

Notice that Scikit-Learn metrics functions always take the true labels and the predicted labels as arguments in that order.

The benefits of MAE are that it is simple to calculate and easy to interpret. It penalizes all errors proportionally based on the magnitude of the error. For example, an error of 100 increases the MAE twice as much as an error of 50.

##Mean Squared Error (MSE)
Mean squared error is similar to mean absolute error, but instead of taking the absolute values of the errors, it squares the errors instead.

We prefer to have a model that is a little wrong for all of the data vs. a model that is very wrong for only some of the data.

Mean squared error penalizes large errors more because the errors are being squared.

For example, An error of 100 would be a squared error of 100^2 (10,000) whereas an error of 50 would be a squared error of 50^2 (2500). So an error twice as big results in a penalty that is 4 times as much.

Once again, a lower MSE is better as we want to reduce the size of errors, especially large errors.

## MSE in Python

we can calulate mse in python using numpy

```python
# Calculating MSE with numpy
train_MSE = np.mean((y_pred_train - y_train)**2)
test_MSE = np.mean((y_pred_test - y_test)**2)
print(f'Model Training MSE: {train_MSE:,.2f}')
print(f'Model Testing MSE: {test_MSE:,.2f}')
```

Or we can use the Scikit-Learn function in the metrics library.

```python
# Calculating MSE with sklearn
from sklearn.metrics import mean_squared_error
train_MSE = mean_squared_error(y_train, y_pred_train)
test_MSE = mean_squared_error(y_test, y_pred_test)
print(f'Model Training MSE: {train_MSE:,.2f}')
print(f'Model Testing MSE: {test_MSE:,.2f}')
```

The benefit of MSE is that it more dramatically penalizes large errors. It rewards a model (with a lower metric) that is more consistent and mostly avoids large errors. The downside of MSE is that it is hard to interpret. The original target in this example was in dollars, but the MSE is in dollars squared.

If we want a metric that both punishes large errors and returns a more interpretable result, we need yet another metric.

## Root Mean Squared Error (RMSE)

Root mean squared error is the square root of the mean squared error.

Like the previous two metrics, a lower RMSE is better. The benefits of RMSE are that it both punishes large errors (like MSE) and is it is easier to interpret because it is expressed in the same units as the target. By taking the square root of MSE, the units go back to the original units of the target. For example, the square root of dollars^2 will get us back to just dollars which is an easy-to-interpret value.

##RMSE in Python

We can calculate RMSE from scratch using NumPy :




```python
# Calculating RMSE with numpy
train_RMSE = np.sqrt(np.mean((y_pred_train - y_train)**2))
test_RMSE = np.sqrt(np.mean((y_pred_test - y_test)**2))
print(f'Model Training RMSE: {train_RMSE:,.2f}')
print(f'Model Testing RMSE: {test_RMSE:,.2f}')

```

-  There is no Scikit-Learn function to calculate RMSE, but we can use sklearn's mean_squared_error and add squared=False to get the RMSE.

```python
# Calculating RMSE with sklearn
train_RMSE = mean_squared_error(y_train, y_pred_train, squared=False)
test_RMSE = mean_squared_error(y_test, y_pred_test, squared=False)
print(f'Model Training RMSE: {train_RMSE:,.2f}')
print(f'Model Testing RMSE: {test_RMSE:,.2f}')

```

In this case, we can see that our RMSE is significantly larger than the MAE we calculated previously. This tells us that while our model is making an average error (MAE) of $19,928 on the test data, it is making some much larger errors! The RMSE is punishing those larger errors with a worse score.

RMSE is a common reporting metric for regression models. In most cases, large errors are disproportionately much worse than small ones, and we would prefer a model that consistently makes a higher average error rather than a model that makes a lower average error but makes some very large errors. Remember that in real-world deployment, we cant predict the error of a prediction until it is generally too late to do anything about it. These unexpected large errors can be very bad compared to consistent expected errors.

MAE, MSE, and RMSE scores are all dependent on the scale and units of the target. This means that you can't interpret any of these errors without context! There is no particular value for MAE, MSE, or RMSE that means anything out of context. An MSE of 5,000 could be good or bad, depending on what it is referring to! Being off by $5000 on the sale of a house isn't too bad, but being off by $5000 on the sale of a car would be horrible! But what if we want one metric that is on the same scale for all datasets? Theres one more that we will talk about here.

##R-Squared (R2, R^2)


 the "coefficient of determination," describes the percentage of the variation in the target variable that a model can explain by using all the features together.

So an R^2 of .80 means that the model can explain 80% of the variation in the target.

Whereas for the previous metrics, a lower score was better, a higher R2 score is better. A score of 1 means a perfect fit on the data and a score of 0 means that the model does not predict any of the variance of the target. An R2 score can be negative, but this often means there is a serious problem with the model.
##Calculating R-squared
R^2 is calculated by determining the correlation between the actual values and the predictions made by the model (r), and then squaring it to get R^2.
##R-squared in Python
We can calculate the R-squared in NumPy:


```python
# Calculating R-Squared
train_r2 = np.corrcoef(y_train, y_pred_train)[0][1]**2
test_r2 = np.corrcoef(y_test, y_pred_test)[0][1]**2
print(f'Model Training R2: {train_r2:.2f}')
print(f'Model Testing R2: {test_r2:.2f}')
```

Or we can let Scikit-Learn calculate it for us.

```python
from sklearn.metrics import r2_score
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)
print(f'Model Training R2: {train_r2:.2f}')
print(f'Model Testing R2: {test_r2:.2f}')
```

This score can be interpreted as saying that our model can account for about 83% of the variation in y_test using the features in X_test.

The benefits of R2 are that it is on a consistent scale for all datasets. You dont have to refer to the scale of the target or consider the business problem to get a rough sense of how well your model is fitting the data. The downside is that R2 is difficult to interpret and very difficult to explain to non-technical audiences. Also, a high R2 doesnt always mean a good model and a low one doesnt always mean a bad one.

#Bias and variance


Bias and variance are two of the most important ideas in machine learning. These describe how your model is doing in more general terms than specific metrics. A good model has both a low variance and a low bias.

- Bias: How well or poorly a model is able to predict a target using features. A low bias is better.
- Underfitting: Another word for bias. An underfit model makes poor predictions on both a training and testing set.
- Variance: How much a models predictive power will vary using different datasets. A low variance is better.
- Overfitting: Another name for variance. An overfit model makes good predictions on a training set, but poor predictions on a testing set.
- Noise: Natural randomness in data.
- Signal: The true function of features to a target.
## signal and noise


image.png


The assumption of all predictive models is that there is a relationship between features (X) and target (y). This is called the signal. A model tries to find that signal.

However, there is also noise in data. This is randomness caused by factors other than the features in our data. This noise is considered irreducible because it is not part of the true signal. Perhaps if our features included all information in the universe, we could eliminate noise. However, in reality, a real world dataset will always have irreducible noise.

The image above shows the signal in the relationship between x and y. A model with low bias and low variance can find that signal or something close to it. Notice, however, that the signal does not perfectly line up with all of the data. It does, however, minimize the error. The remaining error is irreducible and is called noise.


##Bias/Underfitting

image.png

A model with high bias is consistently unable to make good predictions using any set of features. The model underfit the data. The model has not found the patterns in the data that correlate to the target it is trying to predict.

Identifying Underfitting
You will know your model is underfit when it performs poorly on both the training and testing data.

For instance, when you evaluate the R2 Score of your model on your training and testing set, you find they are both around .35. Or, perhaps the mean absolute error is larger than the mean value of the target column.

Causes of Underfitting
- A model that is too simple (see graphic above).
- Not enough data.
- Features that do not correlate well with the target.

For instance, if you were to create a decision tree model with a max depth of 2 (a simple model) to predict the temperature tomorrow using todays stock prices (features do not correlate with target) and a training set of 5 samples (not enough data), your model would most certainly have high bias.


Notice in the graphic above that a high bias model has not found the signal and has a high error on most data.

Reducing Bias:

  You can combat bias by:

- Increasing the complexity of your model.
- Adding more data.
- Adding features with higher correlation to the target.

##Variance/Overfitting
image.png

A model with high variance can make good predictions using some data, but not others. A model with high variance has overfit on the training data. It fits on the noise that is not a part of the true signal. An overfit model will not be able to make good predictions on new data, such as a test set. Its accurate with some and not with others. A high variance often means that you can improve your models performance on the test set, which is most important.

You will know your model is overfit on the training data and high variance if the model performs much better on the training data then on the testing data.

- Identifying Overfitting
Perhaps the R2 score on the training data is .90 and the R2 score on the testing data is only .5. Or, perhaps the RMSE on the training data is much lower than RMSE on the testing data.

- Causes of Overfitting:
  - Too complex of a model (see graphic above).
  - Not enough data.
  - Training data that is not a representative sample of the population, all new data the model might encounter.
- Notice in the graphic above that the model has fit on each and every data point. It has fit on the noise in the data rather than the signal.

- Combatting Overfitting
  - Decrease the complexity of the model.
  - Add regularization (you will learn more about this in another lesson).
  - Add more data to the training set.
  - Ensure that the training set is a representative subset of all data the model will encounter.

##Bias/Variance Tradeoff


image.png

- If you have been paying attention you will notice that some of the techniques that reduce bias are the cause of variance and vice versa. This is the bias/variance trade-off. When decreasing one you will tend to increase the other.

However, notice that more data tends to decrease both!! More data, as long as it is representative to the whole population of data and contains features that are relevant to the target, is always better for training a model with more predictive power.

##Both Bias and Variance are Bad!
- In your work as a machine learning engineer, you will be striving to balance and decrease both bias and variance. An ideal model makes low-error predictions with both the testing and training data. High bias means your model has not found the signal and high variance means there might be a model that can perform better on the testing data.


###Important Note!
The most important thing to pay attention to in choosing a model is the performance on the testing data.

A high variance model that performs better on testing data is better than a low variance model that performs worse on the testing data.

##Summary

- High bias is underfitting and means a model is not able to make accurate predictions using any dataset.
-  High variance is overfitting and means a model performs well with the data it was trained on, but not new data it hasnt seen before.
 There are several ways to combat bias and variance, but reducing one often increases the other. Its best to try to find a balance. However, good performance on testing data is the most important consideration for choosing a model.

Sklearn has an object called a column selector which is a function that will create a list of features that match a specific dtype.

We create one using the make_column_selector function and indicate which dtype we want to include (or exclude). make_column_selector returns a function that will return a list of column names of that type.

The advantage of using a column_selector instead of saving a list of column names using select_dtypes is that the column_selector can change which features it expects to see in the data after it has been instantiated. This will allow us to have our ColumnTransformer automatically update the columns it expects based on the training data.

##8) Evaluate the Model

We will use the custom functions we created in the "Regression Evaluation Functions" lesson. Make sure you have defined them before calling the functions.

```python
def regression_metrics(y_true, y_pred, label='', verbose = True, output_dict=False):
  # Get metrics
  mae = mean_absolute_error(y_true, y_pred)
  mse = mean_squared_error(y_true, y_pred)
  rmse = mean_squared_error(y_true, y_pred, squared=False)
  r_squared = r2_score(y_true, y_pred)
  if verbose == True:
    # Print Result with Label and Header
    header = "-"*60
    print(header, f"Regression Metrics: {label}", header, sep='\n')
    print(f"- MAE = {mae:,.3f}")
    print(f"- MSE = {mse:,.3f}")
    print(f"- RMSE = {rmse:,.3f}")
    print(f"- R^2 = {r_squared:,.3f}")
  if output_dict == True:
      metrics = {'Label':label, 'MAE':mae,
                 'MSE':mse, 'RMSE':rmse, 'R^2':r_squared}
      return metrics

def evaluate_regression(reg, X_train, y_train, X_test, y_test, verbose = True,
                        output_frame=False):
  # Get predictions for training data
  y_train_pred = reg.predict(X_train)

  # Call the helper function to obtain regression metrics for training data
  results_train = regression_metrics(y_train, y_train_pred, verbose = verbose,
                                     output_dict=output_frame,
                                     label='Training Data')
  print()
  # Get predictions for test data
  y_test_pred = reg.predict(X_test)
  # Call the helper function to obtain regression metrics for test data
  results_test = regression_metrics(y_test, y_test_pred, verbose = verbose,
                                  output_dict=output_frame,
                                    label='Test Data' )

  # Store results in a dataframe if ouput_frame is True
  if output_frame:
    results_df = pd.DataFrame([results_train,results_test])
    # Set the label as the index
    results_df = results_df.set_index('Label')
    # Set index.name to none to get a cleaner looking result
    results_df.index.name=None
    # Return the dataframe
    return results_df.round(3)
```

##Finding the Optimal max_depth (hyperparameter tuning)

First, let's manually try a few options to see how changing this parameter affects our model.

To illustrate how changing a hyperparameter can impact the results, we will start by setting the max depth to 2.

Note that we have combined instantiation, fitting, predicting, and evaluating into one code cell. When we instantiate the model, we include the argument max_depth = 2

```python
# Let's go extreme and set our max depth to 2
dec_tree_2 = DecisionTreeRegressor(max_depth = 2, random_state = 42)
# Fit on training data
dec_tree_2.fit(X_train_tf, y_train)
# Get predictions
train_preds = dec_tree_2.predict(X_train_tf)
test_preds = dec_tree_2.predict(X_test_tf)
# Calculate r2 score
train_2_score = r2_score(y_train, train_preds)
test_2_score = r2_score(y_test, test_preds)
# Print results
print (f'Training R2 is {train_2_score:.3f}.')
print (f'Testing R2 is {test_2_score:.3f}.')
```

Let's try a value somewhere between 2 and 27. We chose to try a max depth of 10:

```python
# Instantiate model with max_depth of 10
dec_tree_10 = DecisionTreeRegressor(max_depth = 10, random_state = 42)
# Fit on training data
dec_tree_10.fit(X_train_tf, y_train)
# Get predictions
train_preds = dec_tree_10.predict(X_train_tf)
test_preds = dec_tree_10.predict(X_test_tf)
# Calculate r2 score
train_10_score = r2_score(y_train, train_preds)
test_10_score = r2_score(y_test, test_preds)
# Print results
print (f'Training R2 is {train_10_score:.3f}.')
print (f'Testing R2 is {test_10_score:.3f}.')

```

These results are an improvement, but you can quickly see that manually selecting parameters is time-consuming. We can use a loop to test multiple values more efficiently.

We will make a loop to test all values for max_depth between 2 and 27.

```python
# List of values to try for max_depth:
depths = list(range(2, 27)) # will try every value between 2 and 27
# list to collect the results of each depth
scores_list = []
# Loop through depths
for depth in depths:
    dec_tree = DecisionTreeRegressor(max_depth=depth, random_state=42)
    dec_tree.fit(X_train_tf, y_train)
    # Get predictions
    train_preds = dec_tree.predict(X_train_tf)
    test_preds = dec_tree.predict(X_test_tf)
    # Calculate score
    train_r2 = r2_score(y_train, train_preds)
    test_r2 = r2_score(y_test, test_preds)
    # Add dictionray of scores to score_list
    results = {'max_depth':depth,
               'Train R2':train_r2,
               'Test R2':test_r2}
    scores_list.append(results)
    # Print the results
    print(f'When max_depth is {depth},')
    print(f'- the training R2 is {train_r2:.3f}.')
    print(f'- the testing R2 is {test_r2:.3f}.')
    print()
```

We can put these results into a dataframe.

```python
# Create a dataframe from the scores_list
scores = pd.DataFrame(scores_list)
# Set the max_depth as the index
scores = scores.set_index('max_depth')
scores.head()
```

```python
# Let's run the model with our optimized value for max_depth
dec_tree_4 = DecisionTreeRegressor(max_depth = 4, random_state = 42)
# Fit on the training data
dec_tree_4.fit(X_train_tf, y_train)
# Get predictions
train_preds = dec_tree_4.predict(X_train_tf)
test_preds = dec_tree_4.predict(X_test_tf)
# Calculate score
train_4_score = r2_score(y_train, train_preds)
test_4_score = r2_score(y_test, test_preds)
# Print results
print (f'Training R2 is {train_4_score:.3f}.')
print (f'Testing R2 is {test_4_score:.3f}.')

```

## What is CV

CV stands for cross-validation. This is an enhanced validation technique building off a simple train test split.

image.png

After the training and testing data are separated as usual, the testing data is set aside and only used for the final evaluation (after the tuning process). Throughout the tuning process, only the training data is used.

However, the training data is divided into groups or "folds." By default, GridSearchCV uses 5 folds, as shown in the image above.

Each model will be fit and evaluated multiple times (once for each fold).

Each time, a different group (fold) of the data is used as the testing set.

For example, the first time the model is fit and evaluated, Fold 1 is used for testing. Then the same model is fit and evaluated again, but this time Fold 2 is used for testing. For 5-fold cross-validation, this will happen 5 times.

The scores are then averaged together to get the overall score for that particular model.

Multiple models are tested this way, and the best-performing model will be used for the final evaluation.

The final evaluation uses the entire (no folds) training set as the training data and the preserved test data as the true unseen data.

GridSearchCV
GridSearchCV takes a model type and a dictionary of hyperparameters and ranges of values for each hyperparameter. By default, it will perform a 5-fold cross-validation on a model using each possible combination of hyperparameters values specified in the parameter grid dictionary.

Then GridSearchCV can automatically select the best model, based on a given scoring function, to return.

```python
def regression_metrics(y_true, y_pred, label='', verbose = True, output_dict=False):
  # Get metrics
  mae = mean_absolute_error(y_true, y_pred)
  mse = mean_squared_error(y_true, y_pred)
  rmse = mean_squared_error(y_true, y_pred, squared=False)
  r_squared = r2_score(y_true, y_pred)
  if verbose == True:
    # Print Result with Label and Header
    header = "-"*60
    print(header, f"Regression Metrics: {label}", header, sep='\n')
    print(f"- MAE = {mae:,.3f}")
    print(f"- MSE = {mse:,.3f}")
    print(f"- RMSE = {rmse:,.3f}")
    print(f"- R^2 = {r_squared:,.3f}")
  if output_dict == True:
      metrics = {'Label':label, 'MAE':mae,
                 'MSE':mse, 'RMSE':rmse, 'R^2':r_squared}
      return metrics

def evaluate_regression(reg, X_train, y_train, X_test, y_test, verbose = True,
                        output_frame=False):
  # Get predictions for training data
  y_train_pred = reg.predict(X_train)

  # Call the helper function to obtain regression metrics for training data
  results_train = regression_metrics(y_train, y_train_pred, verbose = verbose,
                                     output_dict=output_frame,
                                     label='Training Data')
  print()
  # Get predictions for test data
  y_test_pred = reg.predict(X_test)
  # Call the helper function to obtain regression metrics for test data
  results_test = regression_metrics(y_test, y_test_pred, verbose = verbose,
                                  output_dict=output_frame,
                                    label='Test Data' )

  # Store results in a dataframe if ouput_frame is True
  if output_frame:
    results_df = pd.DataFrame([results_train,results_test])
    # Set the label as the index
    results_df = results_df.set_index('Label')
    # Set index.name to none to get a cleaner looking result
    results_df.index.name=None
    # Return the dataframe
    return results_df.round(3)

```

###Notes on Hyperparameter Searches
Searching for just the right hyperparameters can take time, sometimes a lot of time! While some hyperparameters should always be tuned for each model you try, some hyperparameters will make only small differences. Your data and your base model choice will usually be more impactful than your hyperparameter tuning.

1. First, work on data.

The choices you make in preparing your dataset, such as how you choose to clean the data, which columns you include, and any feature engineering you perform, will usually have the greatest impact and will generally be where most machine learning engineers spend most of their time.

2. Second, pick the right model type

Your choice of model type, linear models, clustering models, tree models, ensemble models, etc., will generally have the next biggest impact, and its a good idea to try default versions of many model types before starting a hyperparameter search. Again, the exception would be to do some tuning on the few hyperparameters mentioned above when trying those models. Generally, the models that perform the best with their default settings will also perform the best after tuning.

3. Finally, tune hyperparameters.

Hyperparameter tuning, while often very time-intensive, will also often only improve performance a small amount. This should be the last step in your modeling process and should be reserved for your best 1-3 default models with your best versions of your data.

###Additional Options for Tuning
Another tool that is popular among many data scientists is sklearn.model_selection.RandomizedSearchCV. Where GridSearchCV explores all possible combinations of hyperparameter values in your parameter grid, RandomizedSearchCV explores a random sampling of combinations. Many data scientists feel that the EXACT right combination is less important than an approximately best set of hyperparameters. You might choose to experiment with RandomizedSearchCV, especially if you expect an exhaustive search over hyperparameter values will take too long.

###Summary
Tuning hyperparameters of a model can improve performance. sklearn.model_selection. GridSearchCV can speed up and simplify the search for optimal hyperparameter settings. While tuning hyperparameters can improve your model performance, other work can be a better use of limited time, such as exploring data cleaning and preparation options and testing more model types. Hyperparameter tuning should be the last step in your process after your data preparation strategy and your base model are chosen.

##Create Pipelines for Transformation

Note that this is the same code we have been using for preprocessing the Ames data.



- numeric pipeline

```python
# PREPROCESSING PIPELINE FOR NUMERIC DATA
# Save list of column names
num_cols = X_train.select_dtypes("number").columns
print("Numeric Columns:", num_cols)
# instantiate preprocessors
impute_median = SimpleImputer(strategy='median')
scaler = StandardScaler()
# Make a numeric preprocessing pipeline
num_pipe = make_pipeline(impute_median, scaler)
# Making a numeric tuple for ColumnTransformer
num_tuple = ('numeric', num_pipe, num_cols)
```

- ordinal pipeline

```python
# Save list of column names
ord_cols = ['Exter Qual','Exter Cond', 'Garage Qual',"Garage Cond"]
print("Ordinal Columns:", ord_cols)
# Create imputer for ordinal data
impute_na_ord = SimpleImputer(strategy='constant', fill_value='NA')
# Making the OrdinalEncoder
# Specifying order of categories for our  Ordinal Qual/Cond Columms
qual_cond_order = ['NA','Po', 'Fa', 'TA', 'Gd', 'Ex']
# Making the list of order lists for OrdinalEncoder
ordinal_category_orders = [qual_cond_order, qual_cond_order,
                           qual_cond_order, qual_cond_order]
ord_encoder = OrdinalEncoder(categories=ordinal_category_orders)
# Making a final scaler to scale category #'s
scaler_ord = StandardScaler()
# Making an ord_pipe
ord_pipe = make_pipeline(impute_na_ord, ord_encoder, scaler_ord)
# Making an ordinal_tuple for ColumnTransformer
ord_tuple = ('ordinal', ord_pipe, ord_cols)
```

- categorical pipeline

```python
# PREPROCESSING PIPELINE FOR ONE-HOT-ENCODED DATA
# Save list of column names
ohe_cols = X_train.select_dtypes('object').drop(columns=ord_cols).columns
print("OneHotEncoder Columns:", ohe_cols)
# Instantiate the individual preprocessors
impute_na = SimpleImputer(strategy='constant', fill_value = "NA")
ohe_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
# Make pipeline with imputer and encoder
ohe_pipe = make_pipeline(impute_na, ohe_encoder)
# Making a ohe_tuple for ColumnTransformer
ohe_tuple = ('categorical', ohe_pipe, ohe_cols)
```

- Instantiate the Column Transformer

```python
# Create the Column Transformer
preprocessor = ColumnTransformer([num_tuple, ord_tuple, ohe_tuple],
                                     verbose_feature_names_out=False)
```

Remember, we won't fit the transformer yet. We will fit the preprocessor and the model in one step once it is in the pipeline.

##Serialization

In order to save and reuse/deoply a previously trained model, the model must be converted into a form that can be saved and loaded again. This process is called "serialization."

"Serialization is the process of converting a data structure or object into a data format that can be stored . . . or transmitted . . . and reconstructed later. . ." - Serialization (Wikipedia)

Python comes with a built-in module for serialization called pickle, which can save objects as .pickle files. However, it is not the method that is recommended by scikit-learn.

According to scikit-learn's Model Persistence page from the User Guide: "In the specific case of scikit-learn, it may be better to use joblibs replacement of pickle (dump & load), which is more efficient on objects that carry large numpy arrays internally as is often the case for fitted scikit-learn estimators. . ."

As data scientists, our primary use case for serialization is for saving models and preprocessing tools. We will therefore use the tool recommended by scikit-learn: joblib.
